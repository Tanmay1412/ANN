```python
import numpy as np
```

```python
X = np.array([[1,1,1],
              [-1,-1,-1]])
```

```python
Y = np.array([[1,1],
              [-1,-1]])
```

```python
W = np.dot(X.T, Y)
```

```python
def sign(x):
    return np.where(x >= 0, 1, -1)
```

```python
for i in range(len(X)):
    x = X[i]
    for _ in range(5):
        y = sign(np.dot(x, W))
        x = sign(np.dot(y, W.T))
    print(y)
```

```python
for i in range(len(Y)):
    y = Y[i]
    for _ in range(5):
        x = sign(np.dot(y, W.T))
        y = sign(np.dot(x, W))
    print(x)
```
