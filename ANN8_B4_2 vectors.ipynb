```python
import numpy as np
```

```python
X = np.array([
    [1, -1, 1, -1],
    [-1, 1, -1, 1]
])
```

```python
n = X.shape[1]
W = np.zeros((n, n))

for p in X:
    W += np.outer(p, p)

np.fill_diagonal(W, 0)

print(W)
```

```python
def sign(x):
    return np.where(x >= 0, 1, -1)
```

```python
def hopfield(x, W, steps=10):
    for _ in range(steps):
        x = sign(np.dot(W, x))
    return x
```

```python
for i in range(len(X)):
    print("Input:", X[i])
    print("Output:", hopfield(X[i], W))
```

```python
noisy = np.array([1, -1, -1, -1])

print("Noisy Input:", noisy)
print("Recovered Output:", hopfield(noisy, W))
```
